## Data Analysis

```python
#univariate analysis # analyzing each single column 
#Price Distribution
sns.histplot(data= df, x = 'price')
```

```python
# Since it is hard to descripe the histogram, it may
#be a good idea to check for outliers using boxplot

sns.boxplot(data=df, x = 'price')
```

```python
# the outliers which is somehow causing the problem 
#must be removed to have a better idea of the distribution
#of price so we first filter out the price by keeping the values
#less than 1500 as thev previous boxplot suggested and then 
#we try to plot it again:
data = df[df['price'] < 1500]
sns.boxplot(data=data, x= 'price')
```

```python
#after removing outliers, we can check the histiogram again:
plt.figure(figsize=(8,5))
sns.histplot(data=data, x= 'price', bins = 100)
plt.ylabel('Frequency')
```

```python
data.columns
```

```python
#Availability distribution

plt.figure(figsize = (6,3))
sns.histplot(data=data, x = 'availability_365')
plt.title('availability_365 Distribution')
plt.ylabel('frequency')
plt.show()
```

```python
data.dtypes
```

```python
#checking average price for each neighborhood
data.groupby(by= 'neighbourhood_group')['price'].mean()
```

```python
#price per bed in each neighbourhood
data.groupby(by= 'neighbourhood_group')['price per bed'].mean()
```

```python
#price per bed
data['price per bed']=data['price']/data['beds']
data.head()
```

```python
# Check correlation of one variable wioth others(numerical column
correl = data[['latitude','longitude', 'price',
'minimum_nights','number_of_reviews','reviews_per_month',
'calculated_host_listings_count','availability_365' ,                
'number_of_reviews_ltm','beds','price per bed']].corr()

correl
```
